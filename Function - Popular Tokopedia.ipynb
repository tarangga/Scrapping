```python
from selenium import webdriver
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome(executable_path=r".\chromedriver.exe")

driver.get("https://www.tokopedia.com/p/handphone-tablet/aksesoris-handphone?page=2&ob=5")
driver.implicitly_wait(5)
driver.execute_script("window.scrollTo(0, 1000)") 
driver.implicitly_wait(5)
product = driver.find_elements_by_xpath('//div[@class="css-bk6tzz e1nlzfl3"]')
len(product)
```

```python
def get_data(el):
    data = {}
    try:
        data['label'] = el.find_element_by_xpath('.//span[@class="css-1r5pcfx"]').text
    except:
        data['label'] = None
        
    data['Name'] = el.find_element_by_xpath('.//span[@class="css-1bjwylw"]').text
    
    try:
        data['discount'] = el.find_element_by_xpath('.//div[@class="css-rn1hus"]/div[1]').text
        data['real_price'] = el.find_element_by_xpath('.//div[@class="css-rn1hus"]/div[2]').text
        data['price'] = el.find_element_by_xpath('.//span[@class="css-o5uqvq"]').text
    except:
        data['discount'] = '0%'
        data['price'] = el.find_element_by_xpath('.//span[@class="css-o5uqvq"]').text
        data['real_price'] = data['price']
    data['location'] = el.find_element_by_xpath('.//div[@class="css-vbihp9"]/span[1]').text
    data['shop'] = el.find_element_by_xpath('.//div[@class="css-vbihp9"]/span[2]').text

    data['num_review'] = el.find_element_by_xpath('.//div[@class="css-153qjw7"]').text[1:-1]
    data['link'] = el.find_element_by_xpath('.//a').get_attribute("href")
    return data
```
