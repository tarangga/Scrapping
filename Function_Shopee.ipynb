```python
def get_fromid(match_id, page, category):
  """
  category is pop, sales, relevancy
  """
  url = 'https://shopee.co.id/api/v4/search/search_items?by={category}&limit=60&match_id={match_id}&newest={page}&order=desc&page_type=search&scenario=PAGE_OTHERS&version=2'.format(
      category=category,
      match_id=match_id,
      page=page*60
  )
  return get(url, headers=headers).json()

def get_item(itemid, shopid):
  url = 'https://shopee.co.id/api/v4/item/get?itemid={itemid}&shopid={shopid}'.format(
      itemid=itemid,
      shopid=shopid
  )
  return get(url).json()

def get_shop(id):
    url = 'https://shopee.co.id/api/v4/product/get_shop_info?shopid={id}'.format(id=id)
    return get(url).json()
```

```python
res = get_fromid(11044458, 0, 'pop')
res = [d['item_basic'] for d in res['items']]
df = pd.DataFrame(res)
df.head()
```
