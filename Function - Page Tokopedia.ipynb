```python
import pandas as pd
from selenium import webdriver
import re
from urllib.parse import unquote
```

```python
df = pd.read_csv('tokopedia.csv')
```

```python
def get_info(link):
    driver = webdriver.Chrome(executable_path=r".\chromedriver.exe")
    print(link)
    try:
        link = 'https://' + unquote(re.search(r'(www\.tokopedia\.com.+)topads', link).group())
    except:
        None
    driver.get(link)
    driver.execute_script("window.scrollTo(0, 500)") 
    driver.implicitly_wait(2)
    driver.execute_script("window.scrollTo(0, 750)") 
    driver.implicitly_wait(3)
    driver.execute_script("window.scrollTo(0, 1000)") 
    driver.implicitly_wait(3)
    data = {}
    data['product'] = driver.find_element_by_xpath('//h1[@class="css-1wtrxts"]').text
    data['num_sale'] = driver.find_element_by_xpath('//div[@class="css-gbp2g2"]/div/div[1]').text[8:]
    details = driver.find_elements_by_xpath('//li[@class="css-1dmo88g"]')
    print(details)
    details = [d.text.split(':') for d in details]
    for k, v in details:
        data[k.strip()] = v.strip()
    
    
    try:
        data['num_review'] = driver.find_element_by_xpath('//div[@class="css-gbp2g2"]/div/div[2]/span[2]').text[1:-8]
    except:
        data['num_review'] = '0'
        
    try:
        data['num_discuss'] = driver.find_element_by_xpath('//div[@class="css-gbp2g2"]/div/div[3]').text[9:-1]
    except:
        data['num_discuss'] = '0'
        
    data['desription'] = driver.find_element_by_xpath('//div[@role="tabpanel"]/div').text
    data['store'] = driver.find_element_by_xpath('//a[@class="css-1n8curp"]').text
    
    stars = driver.find_elements_by_xpath('//div[@class="css-udfbf8 e1ufc1ph0"]')
    data['stars'] = [s.text for s in stars]
    driver.close()
    return data

```

```python
items = list(df.to_dict(orient="index").values())
```

```python
i = 0
fail = []
for item in items[i:]:
    print('\n', i, '\n')
    item['information'] = get_info(item['link'])
    if len(item['information']['stars']) == 0:
        print(i, '----------------------------------------------------------------------------------------------------------')
        fail.append(i)
    i = i + 1
```
